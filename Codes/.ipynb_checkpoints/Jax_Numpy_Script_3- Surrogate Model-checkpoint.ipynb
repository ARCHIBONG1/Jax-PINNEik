<a id="model"></a>
#### Surrogate Velocity Model

```python
# Velocity at the origin (0,0)
v0 = 2; 
z_grad = 0.4; # Vertical gradient
x_grad = 0.1; # Horizontal gradient

z_min = 0.; z_max = 6.; z_delta = 0.05;
x_min = 0.; x_max = 6.; x_delta = 0.05;

# Creating a mesh of sources
sz = jnp.array([0.1,2.1,4.0,5.9])
sx = jnp.array([0.1,2.1,4.0,5.9])

SZ,SX = jnp.meshgrid(sz,sx)

sz = SZ.reshape(-1,1)
sx = SX.reshape(-1,1)
```

```python
# Generating the vertical and horizontal values of our model
z = jnp.arange(z_min,z_max+z_delta,z_delta)
nz = z.size
x = jnp.arange(x_min,x_max+x_delta,x_delta)
nx = x.size
```

```python
Z,X = jnp.meshgrid(z,x,indexing='ij')
```

```python
# Velocity model for plotting
velmodel = v0 + z_grad*(Z-0.) + x_grad*(X-0.);
```

```python
# Model with the source location

plt.style.use('default')         # Sets the default style for the plot

plt.figure(figsize=(4,4))        # Creates a new figure with a size of 4x4 inches

plt.title("Surrogate Velocity Model")    # Sets the title of the figure

ax = plt.gca()                   # Gets the current axes instance.
im = ax.imshow(velmodel, extent=[x_min,x_max,z_max,z_min], aspect=1, cmap="jet") # Plots the velocity model velmodel as an image on the current axes (ax)

ax.plot(sx,sz,'k*',markersize=8) # Plots a black star marker at the source location 

plt.xlabel('Offset (km)', fontsize=14)     # Sets the label for the x-axis
plt.xticks(fontsize=12)          # Sets the font size for the x-axis tick labels

plt.ylabel('Depth (km)', fontsize=14)      # Sets the label for the y-axis
plt.yticks(fontsize=12)          # Sets the font size for the y-axis tick labels

ax.xaxis.set_major_locator(plt.MultipleLocator(2))   # Sets the major tick locator for the x-axis to have ticks at intervals of 0.5.
ax.yaxis.set_major_locator(plt.MultipleLocator(2))   # Sets the major tick locator for the y-axis to have ticks at intervals of 0.5.

divider = make_axes_locatable(ax)         # Creates a divider to make room for the colorbar
cax = divider.append_axes("right", size="6%", pad=0.15) # Appends an axes on the right side of the plot to accommodate the colorbar

cbar = plt.colorbar(im, cax=cax) # Adds a colorbar to the plot using the image im and the colorbar axes cax

cbar.set_label('km/s',size=10)   # Sets the label for the colorbar.
cbar.ax.tick_params(labelsize=10)        # Sets the font size of the colorbar tick labels.

plt.savefig("/home/archibong/Desktop/PINNs/Jax/Archibong1/PINNEik Using Jax/Images/Surrogate Velocity Model.pdf", format= "pdf",bbox_inches="tight")  # Saves the image as a pdf file at the specified file directory

plt.show()   # Displays the defined image
```

<a id="train"></a>
#### Training Data

```python
# Preparing necessary arrays for the sources

X_star, Z_star, SX_star, SZ_star, T0_star, px0_star, pz0_star, tau_star, V_star = [[] for i in range(9)]
```

```python
for ns, (szi, sxi) in enumerate(zip(sz, sx)):

    vs = v0 + z_grad*szi + + x_grad*sxi # Velocity at source point
    velmodel = vs + z_grad*(Z-szi) + x_grad*(X-sxi);

    T_data = jnp.arccosh(1.0+0.5*(1.0/velmodel)*(1/vs)*(z_grad**2 + x_grad**2)*((X-sxi)**2 + (Z-szi)**2))/jnp.sqrt(z_grad**2 + x_grad**2)

    T0 = jnp.sqrt((Z-szi)**2 + (X-sxi)**2)/vs; 
    #px0 = jnp.divide(X-sxi, T0*vs**2, out=jnp.zeros_like(T0), where=T0!=0)
    #pz0 = jnp.divide(Z-szi, T0*vs**2, out=jnp.zeros_like(T0), where=T0!=0)

    px0 = jnp.where(T0 != 0, jnp.divide(X - sxi, T0 * vs**2), 0)
    pz0 = jnp.where(T0 != 0, jnp.divide(Z - szi, T0 * vs**2), 0)

    #tau_data = jnp.divide(T_data, T0, out=jnp.ones_like(T0), where=T0!=0)

    tau_data = jnp.where(T0 != 0, jnp.divide(T_data, T0), 1)

    X_star.append(X.reshape(-1, 1))
    Z_star.append(Z.reshape(-1, 1))
    SX_star.append(jnp.ones_like(X.reshape(-1, 1)).reshape(-1,1)*sxi)
    SZ_star.append(jnp.ones_like(X.reshape(-1, 1)).reshape(-1,1)*szi)
    T0_star.append(T0.reshape(-1,1))
    px0_star.append(px0.reshape(-1,1))
    pz0_star.append(pz0.reshape(-1,1))
    tau_star.append(tau_data.reshape(-1,1))
    V_star.append(velmodel.reshape(-1,1))
```

```python
X_star = jnp.concatenate(X_star, axis=0)
Z_star = jnp.concatenate(Z_star, axis=0)
SX_star = jnp.concatenate(SX_star, axis=0)
SZ_star = jnp.concatenate(SZ_star, axis=0)
tau_star = jnp.concatenate(tau_star, axis=0)
T0_star = jnp.concatenate(T0_star, axis=0)
px0_star = jnp.concatenate(px0_star, axis=0)
pz0_star = jnp.concatenate(pz0_star, axis=0)
V_star = jnp.concatenate(V_star, axis=0)
```

<a id="archi"></a>
#### Network Architecture & Random Initialisation

```python
# defines the number of hidden layers and neuron contained within each layer
layers = [4,20,20,20,20,20,20,20,20,20,20,1]
```

```python
# Function to randomly initialise the network parameters using Xavier iniatialisation
def init_params(layers):

  keys = jax.random.split(jax.random.PRNGKey(0),len(layers)-1)
  params = list()      # Defines an empty list to store the parameters

  for key,n_in,n_out in zip(keys,layers[:-1],layers[1:]):   # For loop iterates over the keys, number of input neurons (n_in), and number of output neurons 
    lb = -(1 / jnp.sqrt(n_in)) # lower bound
    ub = (1 / jnp.sqrt(n_in)) # upper bound
    W = lb + (ub-lb) * jax.random.uniform(key,shape=(n_in,n_out))  # Initialises the weight matrix
    B = jax.random.uniform(key,shape=(n_out,))             # Initialises the bias vector 
    params.append({'W':W,'B':B})                           # Adds a dictionary containing the weight matrix W and bias vector B to the params list
  return params        # Returns the initialised parameters
```

```python
# saves the randomly initialised parameters into params
params = init_params(layers)
```

<a id="tau"></a>
#### Neural Network

```python
# Function tau to computes the output of the neural network 
def tau(params,xt,zt,sxt,szt):

  X = jnp.concatenate([xt,zt,sxt,szt],axis=1)

  *hidden,last = params
  for layer in hidden:

    X = jax.nn.tanh(X@layer['W']+layer['B']) # Applying the hyperbolic tangent (tanh) activation function

  return X@last['W'] + last['B'] 
```

<a id="pde"></a>
#### Eikonal PDE Loss

```python
@jax.jit   # Just-in-Time (JIT) decorator to accelerate computation
def pde_loss(params,xt,zt,sxt,szt,T0t,px0t,pz0t,vt):
    
    # Reshapes all input parameters
    zt = zt.reshape(-1, 1)
    xt = xt.reshape(-1, 1)
    sxt = sxt.reshape(-1, 1)
    szt = szt.reshape(-1, 1)
    T0t = T0t.reshape(-1, 1)
    px0t = px0t.reshape(-1, 1)
    pz0t = pz0t.reshape(-1, 1)
    vt = vt.reshape(-1, 1)

    
    # Computes the output of the neural network
    tau_output = tau(params,xt,zt,sxt,szt) 
    
    # Computes the gradient
    dtau_dz = jax.grad(lambda xt,zt: jnp.sum(tau(params, xt,zt,sxt,szt)), argnums=1)(xt,zt)  # Gradient of the tau function with respect to zt
    dtau_dx = jax.grad(lambda xt,zt: jnp.sum(tau(params, xt,zt,sxt,szt)), argnums=0)(xt,zt)  # Gradient of the tau function with respect to xt

    # Computes the combined term
    term1 = T0t * dtau_dz + tau_output * pz0t
    term2 = T0t * dtau_dx + tau_output * px0t
    
    # Computes the PDE loss value using MSE function 
    loss_pde = MSE(0,(jnp.square(term1) + jnp.square(term2) - 1.0 / (vt ** 2))) 
    
    return loss_pde
```

<a id="other"></a>
#### Other Losses

```python
@jax.jit    # Just-in-Time (JIT) decorator to accelerate computation
def MSE(true_tau,pred_tau):
    # Function to compute mean squared error (MSE)
    
    return jnp.mean((true_tau-pred_tau)**2)  # Computes and returns the MSE between true and pred
```

```python
@jax.jit   # Just-in-Time (JIT) decorator to accelerate computation
def tau_loss(params,xt,zt,sxt,szt,tau_star):
    # Function to compute tau loss
    
    # # Reshapes all input parameters
    zt =zt.reshape(-1,1)
    xt =xt.reshape(-1,1)
    sxt = sxt.reshape(-1,1)
    szt = szt.reshape(-1,1)
    
    # Computes the output of the neural network
    tau_output = tau(params,xt,zt,sxt,szt)  
    
    # Computes the tau loss value using MSE function
    tau_loss = MSE(tau_star, tau_output)
    
    
    return tau_loss
```

<a id="combine"></a>
#### Combined Loss Function

```python
@jax.jit    # Just-in-Time (JIT) decorator to accelerate computation
def loss_fun(params,xt,zt,sxt,szt,tau_star, T0t, px0t, pz0t, vt):
    # Function to compute the combined loss from tau and PDE
    
    loss_tau = tau_loss(params,xt,zt,sxt,szt,tau_star)   # Obtains the tau loss value from tau loss function
    loss_pde = pde_loss(params,xt,zt,sxt,szt,T0t,px0t,pz0t,vt) # Obtains the PDE loss value from PDE loss function
    
    return loss_pde + loss_tau # Computes and returns the sum of the boths losses
```

<a id="new"></a>
#### Model Training

```python
# defines an adaptive learning rate that changes at specified epochs
lr = optax.piecewise_constant_schedule(1.e-2,{810:9.7e-3,70000:8e-3}) #(1e-3) #,{500:9e-4})#2:1e-2,500:2e-3}) #(1e-2,{500:2.9e-3})

# Utilises the adam optimiser
optimizer = optax.adam(lr)

# Updates the parameters
opt_state = optimizer.init(params)
```

```python
@jax.jit    # Just-in-Time (JIT) decorator to accelerate computation
def update(opt_state,params,xt,zt,sxt,szt,tau_star, T0t, px0t, pz0t, vt):
    # Function to update the parameters for each gradient descent
    
    # calculates the gradients of loss_fun
    grads=jax.grad(loss_fun,0)(params,xt,zt,sxt,szt,tau_star, T0t, px0t, pz0t, vt)

    # Update params
    updates, opt_state = optimizer.update(grads, opt_state)
    
    # Applies the updated params
    params = optax.apply_updates(params, updates)

    return opt_state,params
```

```python
xt = X_star
zt=Z_star
sxt=SX_star
szt=SZ_star
tau_star =tau_star
T0t=T0_star
px0t=px0_star
pz0t=pz0_star 
vt=V_star
```

```python
%%time
epochs = 1400 
loss_value = [] # List to hold the values of the combines loss
loss_pde = []   # List to hold the values of the PDE loss
loss_tau = []   # # List to hold the values of the tau loss

for epoch in range(epochs):  # For loop to iterate over the stipulates epochs
    opt_state, params = update(opt_state,params,xt,zt,sxt,szt,tau_star, T0t, px0t, pz0t, vt)

    # print loss and epoch info
    if epoch%(1) ==0:
        loss_pp = loss_fun(params,xt,zt,sxt,szt,tau_star, T0t, px0t, pz0t, vt) # Computes the total loss
        loss_value.append(loss_pp)                                    # Appends the total loss value to the declared list
        print(f'Epoch = {epoch},\tloss = {loss_pp:.3e}')              # Prints the total loss values for each epoch

        #Individual losses
        loss_pde.append(pde_loss(params,xt,zt,sxt,szt,T0t,px0t,pz0t,vt)) # Computes and appends the PDE loss
        loss_tau.append(tau_loss(params,xt,zt,sxt,szt,tau_star))                       # Computes and appends the tau loss
```

```python
# Plotting the Loss Curve
plt.semilogy(loss_value, label="Total", linewidth=2) # Plots the total loss
#plt.semilogy(loss_pde, label="PDE", linewidth=2)     # Plots the PDE loss
#plt.semilogy(loss_tau, label="tau", linewidth=2)    # Plots the tau loss

plt.title("Surrogate Velocity Loss", fontsize=14)   # Sets the title of the figure

plt.xlabel("Epochs ", fontsize=15)              # Sets the x-axis label of the figure
plt.ylabel("Loss", fontsize=15,)                # Sets the y-axis label of the figure

plt.grid(color = 'black', linestyle = '--', linewidth = 0.5) # Sets the grid of the figure

plt.legend(loc="upper right")   # Sets the legend of the figure

plt.savefig("/home/archibong/Desktop/PINNs/Jax/Archibong1/PINNEik Using Jax/Images/Surrogate Velocity Loss.pdf", format="pdf", bbox_inches="tight") # Saves the figure

plt.show() # Displays the figure
```
